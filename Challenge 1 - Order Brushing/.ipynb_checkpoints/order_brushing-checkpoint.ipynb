```python
import pandas as pd
from datetime import datetime, timedelta
```

```python
df = pd.read_csv('order_brush_order.csv')
```

```python
df['event_time'] = pd.to_datetime(df['event_time'], format = "%Y-%m-%d %H:%M:%S")
```

```python
test_df['event_time'].max()
```

```python
output = pd.DataFrame(columns = (['shopid','userid']))
```

```python
for shop in df['shopid'].unique():
    blacklist = pd.DataFrame(columns = (['customer', 'orders']))
    
    test_df = df[df['shopid'] == shop].sort_values('event_time').reset_index(drop=False)
    
    start_time = test_df['event_time'][0]
    end_time = start_time + timedelta(hours=1)
    end = test_df['event_time'].max()
    
    while start_time < end:
        test_df[(test_df['event_time'] >= start_time) & (test_df['event_time'] <= end_time)]
        
        unique_orders = test_df[(test_df['event_time'] >= start_time) & (test_df['event_time'] <= end_time)]['orderid'].nunique()
        unique_customers = test_df[(test_df['event_time'] >= start_time) & (test_df['event_time'] <= end_time)]['userid'].nunique()
        
        try:
            con_r = unique_orders / unique_customers
        except:
            start_time = start_time + timedelta(hours=1)
            end_time = end_time + timedelta(hours=1)
            continue
        
        if con_r >= 3:
            res = test_df[(test_df['event_time'] >= start_time) & (test_df['event_time'] <= end_time)]['userid'].value_counts().reset_index()
            res.columns = (['customer','orders'])
            
            blacklist = blacklist.append(res[res['orders'] ==  res['orders'].max()])
            
        start_time = start_time + timedelta(hours=1)
        end_time = end_time + timedelta(hours=1)
    
    to_append = blacklist.groupby('customer').sum().reset_index()
    to_append = to_append[to_append['orders'] == to_append['orders'].max()]
    
    if len(to_append) > 1:
        out_user = to_append['customer'].astype(str).str.cat(sep='&')
    elif len(to_append) == 0:
        out_user = 0
    else:
        out_user = to_append['customer'][0]
        
    output = output.append(pd.DataFrame([[shop, out_user]], columns = output.columns))
    
    print(shop)
    
    
    
```

```python
output.to_csv('output.csv', index = False)
```
